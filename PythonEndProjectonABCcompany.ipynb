Analysis Tasks:

1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

2. Segregate employees based on their positions within the company.

```python
Positions
```

```python
Position_groups=df.groupby('Position')
```

```python
for Position, group in Position_groups:
    print(f"Position: {Position}")
    print(group)
    print()
```

3. Identify the predominant age group among employees.

```python
#Define age ranges
Age_bins=[20,30,40,50,60,70]
Age_groups=pd.cut(df['Age'],bins=Age_bins)
Predominent_age_group=Age_groups.value_counts().idxmax()
```

```python
print("Predominent Age Group:", Predominent_age_group)
```

4. Discover which team and position have the highest salary expenditure.

```python
Team_Salary=df.groupby('Team')['Salary'].sum()
```

```python
Team_Highest_Salary=Team_Salary.idxmax()
```

```python
Team_Highest_Salary
```

```python
Position_Salary=df.groupby('Position')['Salary'].sum()
```

```python
Positional_Highest_Salary=Position_Salary.idxmax()
```

```python
Positional_Highest_Salary
```
